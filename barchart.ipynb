```
import numpy as np
import matplotlib.pyplot as plt
days = ['monday', 'tuesday','wednsday', 'thursday', 'friday', 'saturday','sunday',]
hours = []
for i in range(7):
    hours.append(np.around((np.random.rand() *10), decimals = 0))
plt.pie(hours, labels = days)
plt.scatter(days, hours)
plt.show()
```
